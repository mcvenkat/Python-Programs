Step #4: Data Cleaning /Feature Engineering

```python
alexa_df = alexa_df.drop(['date', 'rating'], axis = 1)
```

```python
alexa_df
```

```python
variation_dummies = pd.get_dummies(alexa_df['variation'], drop_first = True)
```

```python
variation_dummies
```

```python
alexa_df.drop(['variation'], axis = 1, inplace = True)
```

```python
alexa_df = pd.concat([alexa_df, variation_dummies], axis = 1)
```

```python
alexa_df
```

```python
from sklearn.feature_extraction.text import CountVectorizer
vectorizer = CountVectorizer()
alexa_count_vectorizer = vectorizer.fit_transform(alexa_df['verified_reviews'])
```

```python
alexa_count_vectorizer.shape
```

```python
print(vectorizer.get_feature_names())
```

```python
print(alexa_count_vectorizer.toarray())
```

```python
alexa_df.drop(['verified_reviews'], axis =1, inplace = True)
```

```python
alexa_df
```

```python
encoded_reviews = pd.DataFrame(alexa_count_vectorizer.toarray())
```

```python
alexa_df = pd.concat([alexa_df, encoded_reviews], axis = 1)
```

```python
alexa_df
```

```python
X = alexa_df.drop(['feedback'], axis = 1)
```

```python
Y = alexa_df['feedback']
```

```python
X
```

```python
Y
```

Step #5: Model Training

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size =  0.2, random_state = 5)
```

```python
from sklearn.metrics import confusion_matrix, classification_report
from sklearn.ensemble import RandomForestClassifier
randomforest_classifier = RandomForestClassifier(n_estimators = 100, criterion = 'entropy')
randomforest_classifier.fit(X_train, Y_train)
```

Step #6: Evaluate the Model

```python
Y_predict_train = randomforest_classifier.predict(X_train)
```

```python
cm = confusion_matrix(Y_train, Y_predict_train)
```

```python
sns.heatmap(cm, annot= True)
```

```python
print(classification_report(Y_train, Y_predict_train))
```

```python
Y_predict_test = randomforest_classifier.predict(X_test)
```

```python
cm = confusion_matrix(Y_test, Y_predict_test)
```

```python
sns.heatmap(cm,annot= True)
```

```python
print(classification_report(Y_test, Y_predict_test))
```
